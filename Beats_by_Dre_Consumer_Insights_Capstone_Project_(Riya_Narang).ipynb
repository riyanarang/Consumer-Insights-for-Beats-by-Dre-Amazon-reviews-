```python
import pandas as pd
import sklearn as sk
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('amazon_ref_dataset.csv')

# Descriptive statistics
df.describe()['rating']
```

```python
#Extract date using regex with error handling
def extract_date(timestamp):
    if pd.isna(timestamp):
        return None

    # Define a regular expression to capture the date (Month Day, Year)
    match = re.search(r'(\w+\s\d{1,2},\s\d{4})', str(timestamp))

    if match:
        return match.group(0)
    return None


df['extracted_date'] = df['timestamp'].apply(extract_date)
```

```python
# Extracting title from 'title' column and adding it to content column
pattern = r'(?P<stars>.*)'

df['title'] = df['title'].apply(lambda x: re.findall(pattern, str(x))[0].strip() if re.findall(pattern, str(x)) else '')
df['content']= df['title']+ df['content']
```

```python
# Extracting Brand name
df['brand'] = df.product_name.str.split()
df['brand'] = df['brand'].apply(lambda x: x[0])
```

```python
# Extracting color from 'product_attributes' column
df['color'] = df['product_attributes'].str.extract(r'Color:\s*(\w+)')
df['color'].fillna('Unknown', inplace=True)
```

```python
df=df[['review_id', 'product_name', 'brand', 'author', 'rating','content', 'color', 'extracted_date','helpful_count', 'is_verified']]
df.head()
```

```python
import google.generativeai as genai

df = pd.read_csv("amazon_ref_dataset.csv")
def analyze_reviews(reviews, api_key,prompt_message):
  genai.configure(api_key=api_key)

# Create the model
  generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
  }

  model = genai.GenerativeModel(
   model_name="gemini-1.5-flash",
   generation_config=generation_config,
  # safety_settings = Adjust safety settings
  # See https://ai.google.dev/gemini-api/docs/safety-settings
  )

  chat_session = model.start_chat(history=[])
  prompt = prompt_message + "\n\n".join(str(review) for review in reviews)
  response = chat_session.send_message(prompt)

  return response.text
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
df['sentiment_score'] = df['content'].apply(lambda review: analyzer.polarity_scores(review)['compound'])
df[['content', 'sentiment_score']].head()
```

```python
from textblob import TextBlob

def sentiment_analysis_summary(df, text_col):
  positive, negative, neutral = 0, 0, 0

  for review in df[text_col]:
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
      positive += 1
    elif polarity < 0:
      negative += 1
    else:
      neutral += 1

  return {
      'positive': positive,
      'negative': negative,
      'neutral': neutral
  }

sentiment_analysis_summary(df, 'content')
```

```python
sentiment=df.groupby('brand')['sentiment_score'].mean().reset_index()
display(sentiment)
```

### 8.2 Comparison of Sentiments
Compare the sentiment analysis results of Beats by Dre with its competitors using visualizations to highlight differences and similarities.

```python
def get_sentiment_category(review):
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment analysis to each review
df['sentiment'] = df['content'].apply(get_sentiment_category)

# Group by brand and sentiment to count each sentiment type per product
sentiment_counts = df.groupby(['brand', 'sentiment']).size().unstack().fillna(0)

# Calculate sentiment percentages for each brand
sentiment_percentages = sentiment_counts.div(sentiment_counts.sum(axis=1), axis=0) * 100


ax = sentiment_percentages[['Positive', 'Neutral', 'Negative']].plot(
    kind='bar',
    stacked=True,
    figsize=(8, 6))

plt.xlabel('Products')
plt.ylabel('Percentage')
plt.title('Sentiment Breakdown by Brand')


for label in ax.get_xticklabels():
    if 'Beats' in label.get_text():
        label.set_color('red')


plt.legend(title='Sentiment Type')
plt.xticks(rotation=45, ha='right')
plt.show()
```
